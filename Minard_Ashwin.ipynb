```python
import pandas as pd
import numpy as np
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
from plotly import tools
import plotly.offline as py
```

```python
minard = pd.read_excel("minard-data.xlsx", header=0) 
minard
```

```python
#Code for the extension of line
sur1 = minard[((minard['DIR']=='A') & (minard['DIV']==1)) | 
                   ((minard['DIR']=='R') & (minard['DIV']==1) & (minard['SURV']==100000))]
sur2 = minard[((minard['DIR']=='A') & (minard['DIV']==2)) | 
                   ((minard['DIR']=='R') & (minard['DIV']==2) & (minard['LATP']==55.5))]
sur3 = minard[(minard['DIR']=='A') & (minard['DIV']==3)]
sur4 = minard[(minard['DIR']=='R') & (minard['DIV']==2)]
sur5 = minard[(minard['DIR']=='R') & (minard['DIV']==1)]
sur6 = minard[(minard['DIR']=='R') & (minard['DIV']==3) | 
                   (minard['DIR']=='A') & (minard['DIV']==3) & (minard['LATP']==55.3)]
```

```python
traceCities = go.Scatter(
    x = minard.LONC,
    y = minard.LATC,
    showlegend = False,
    mode ='markers+text',
    marker = dict(
        color = "black",
        size = 8
    ),
    text = minard.CITY,
    textposition='middle right'
  )
```

```python
trace1A = go.Scatter(
     x = sur1.LONP,
     y = sur1.LATP,
     name='ADVANCE',
    
     line = dict(
        color = ('#996515'),
        width = 10
        ),
  
     mode='lines'
)
```

```python
trace1R = go.Scatter(
     x= sur5.LONP,
     y= sur5.LATP,
      name='RETREAT', 
  line = dict(
        color = ('lightskyblue'),
        width = 8),
  mode='lines'
  )
```

```python
#Code for extension of trace2A to trace2R
trace2A = go.Scatter(
     x= sur2.LONP,
     y= sur2.LATP,
     showlegend=False,

  line = dict(
        color = ('#996515'),
        width = 10),
  mode='lines',
  textposition='bottom center'

)
```

```python
trace2R = go.Scatter(

     x= sur4.LONP,
     y= sur4.LATP,
     showlegend=False,
    
  line = dict(
        color = ('lightskyblue'),
        width = 8),
  mode='lines',
  text=sur4.CITY,
  )
```

```python
#extension of 3A & 3R
trace3A = go.Scatter(   
    x= sur3.LONP,
    y= sur3.LATP,
    showlegend=False,
  
    line = dict(
        color = ('#996515'),
        width = 10),
        mode='lines',
        textposition='bottom center'
)
```

```python
trace3R = go.Scatter(
    x = sur6.LONP,
    y = sur6.LATP,
    showlegend = False,
    
    line = dict(
        color = ('lightskyblue'),
        width = 8
    ),
  
    mode = 'lines',
    textposition = 'bottom center'
)
```

```python
traceTemp = go.Scatter(
    x = minard.LONT,
    y = minard.TEMP,
    showlegend = False,
    mode = 'lines+text+markers',
    marker = dict(
        symbol = 'square',
        color = "#d62728",
        size = 8
    ),
    text = minard.TEMP.fillna(0).astype(int).map(str) + ' On ' + minard.MON + ' ' 
            + minard.DAY.fillna(0).astype(int).astype(str),
    textposition = 'bottom center'
    
  )
```

```python
# Subplot for temprature
fig = tools.make_subplots(rows=2,cols=1)

fig.append_trace(trace1A,1,1)
fig.append_trace(trace1R,1,1)
fig.append_trace(trace2A,1,1)
fig.append_trace(trace2R,1,1)
fig.append_trace(trace3A,1,1)
fig.append_trace(trace3R,1,1)
fig.append_trace(traceCities,1,1)
fig.append_trace(traceTemp,2,1)


fig['layout'].update(
    title = dict(
        text = 'CHARLES JOSEPH MINARD’S MAP OF NAPOLEON’S RUSSIA CAMPAIGN.',
        font = dict(
            size = 20,
            family = "Droid Serif"
        ),
        xanchor = "center"
    ),
    
    margin = dict(
        b = 50,
        t = 50,
        pad = 5
    ),
    showlegend=True,
    width = 1000,
    height = 800,
    yaxis = dict(visible=False),
    yaxis2 = dict(title= 'Temperature',ticklen= 0.5,zeroline= False)
)

py.iplot(fig, validate = False)
```
